```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/VINEET/Desktop/ICRISAT-District Level Data.csv")
df
```

```python
a=df.drop(["State Name","Dist Name"],axis=1)
a
```

```python
df2=pd.DataFrame(a)
df2
```

```python
x=df2.iloc[:,1:51].values
x.shape
```

```python
y=df2.iloc[:,0].values
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
pred=knn.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

results=confusion_matrix(y_test,pred)
print('confusion matrix')
print(results)
print('Accuracy score:',accuracy_score(y_test,pred))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
x_train_new=pca.fit_transform(x_train)
x_test_new=pca.transform(x_test)
```

```python
print(x_train.shape)
```

```python
print(x_train_new.shape)
```

```python
pca.explained_variance_ratio_
```

```python
knn.fit(x_train_new,y_train)
```

```python
pred_new=knn.predict(x_test_new)
pred_new
```

```python
accuracy_score(y_test,pred_new)
```
